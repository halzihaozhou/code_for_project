```python
# RFM model
# we separate each dimension(R,F,M) into 2 category(high,low), therefore, customers are segmentaed into 8 groups.
```

```python
qry8="SELECT customer_id,MAX(created_at) as order_date, COUNT(id) as F,\
ROUND(SUM(total_price),2) AS M FROM orders WHERE cancelled_at IS NULL GROUP BY 1;"
```

```python
df_RFM=pd.read_sql_query(qry8,engine)
```

```python
df_RFM
```

```python
df_RFM['R']=(df_RFM['order_date'].max()-df_RFM['order_date'])/np.timedelta64(1,'D')
```

```python
df_RFM
```

```python
df_RFM.drop('order_date',axis=1,inplace=True)
```

```python
df_RFM=df_RFM[['customer_id','R','F','M']]
```

```python
df_RFM
```

```python
def RFM_classify(df):
    for i in range(len(df.iloc[:,0])):
            if df.iloc[i,1]<df['R'].mean():
                df.iloc[i,1]=1
            else:
                df.iloc[i,1]=0
            if df.iloc[i,2]<df['F'].mean():
                df.iloc[i,2]=0
            else:
                df.iloc[i,2]=1
            if df.iloc[i,3]<df['M'].mean():
                df.iloc[i,3]=0
            else:
                df.iloc[i,3]=1
```

```python
RFM_classify(df_RFM)
```

```python
df_RFM=pd.DataFrame(df_RFM,dtype=int)
```

```python
df_RFM
```

```python
df_RFM['indicator']=[str(df_RFM.iloc[i,1])+str(df_RFM.iloc[i,2])+str(df_RFM.iloc[i,3])for i in range(len(df_RFM.iloc[:,0]))]
```

```python
df_RFM
```

```python
dic={'111':'Core Customers','011':'Customers at risk',
                             '101':'Potential Loyalist','110':'Loyal Customers',
                             '100':'Promising Customers','010':'Faithful Customer',
                              '001':'Highest paying customer','000':'Sleep Customers'}
df_RFM['customer segment']=df_RFM['indicator'].map(dic)
```

```python
df_RFM
```

```python
# now let's plot customer segmentation
df_plot=pd.DataFrame(df_RFM['customer segment'].value_counts())
df_plot
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=45)
plt.title('Customer Cluster Distribution')
sns.barplot(x=df_plot.index,y='customer segment',data=df_plot,color='orange')
```
